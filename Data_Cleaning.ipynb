```python
# 1. Clean Price Column

def clean_price(x):
    if isinstance(x, str):
        x = x.replace("₹", "").replace(",", "").strip().lower()

        # Match lakh or cr in text
        if "lakh" in x:
            return float(re.sub("[^0-9.]", "", x)) * 100000
        elif "l" in x:
            return float(re.sub("[^0-9.]", "", x)) * 100000
        elif "cr" in x:
            return float(re.sub("[^0-9.]", "", x)) * 10000000
        elif x == "" or x == "nan":
            return np.nan
        else:
            return float(x)
    return x

df['Price'] = df['Price'].apply(clean_price)
```

```python
# 2. Clean EMI_Price Column

def clean_emi(x):
    if isinstance(x, str):
        x = x.replace("EMI", "").replace("₹", "").replace(",", "").replace("/m", "").strip()
        if x == "" or x.lower() == "nan":
            return np.nan
        return float(x)
    return x

df['EMI_Price'] = df['EMI_Price'].apply(clean_emi)

# Fill Missing EMI with median
df['EMI_Price'] = df['EMI_Price'].fillna(df['EMI_Price'].median())
```

```python
# 3. Standardize Text Columns

text_cols = ['City', 'Brand', 'Model', 'Fuel_Type', 'Transmission', 'Owner']
for col in text_cols:
    df[col] = df[col].astype(str).str.title().str.strip()
```

```python
# 4. Derived Columns

# Car Age
current_year = datetime.datetime.now().year
df['Car_Age'] = current_year - df['Year']

# Price per KM
df['Price_per_KM'] = (df['Price'] / df['KM_Driven']).round(2)

# Price Category
df['Price_Category'] = pd.cut(
    df['Price'],
    bins=[0, 300000, 700000, 1500000, 5000000],
    labels=['Budget', 'Mid-Range', 'Premium', 'Luxury']
)

# KM Representation (K = Thousands, L = Lakhs)
df['KM_Representation'] = df['KM_Driven'].apply(
    lambda x: 'L' if x >= 100000 else ('K' if x >= 1000 else np.nan)
)
```
